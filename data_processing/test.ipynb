```python
import cv2
import glob
import matplotlib.pyplot as plt
```

```python
def view(lb):
    path = "/datahdd/nhanv/git/VAIPE_AI/datasets/VAIPE_PILL_CLS/train/"+str(lb)+"/*.jpg"
    images=[image for image in glob.glob(path)]
    images = images[0:9]
    fig = plt.figure()
    for i in range(len(images)):
        ax = fig.add_subplot(3,3,i+1)
        image = plt.imread(images[i])
        ax.imshow(image)

    plt.show()
```

```python
image_public_test_name = "VAIPE_P_166_12.jpg"
image = plt.imread("/datahdd/public_test/pill/image/"+image_public_test_name)
plt.imshow(image)
plt.show()
```

```python
image_pres_test_name = "VAIPE_P_TEST_74.png"
image = plt.imread("/datahdd/public_test/prescription/image/"+image_pres_test_name)
plt.imshow(image)
plt.show()
```
